```python
import numpy as np
import matplotlib.pyplot as pyplot 


jastrow = np.loadtxt("./results/energy_jastrow_matrix.txt")

jsx = jastrow[:, 0]
jsy = jastrow[:, 1]

rbm = np.loadtxt("./results/energy_rbm.txt")

rbmx = rbm[:, 0]
rbmy = rbm[:, 1]

ngd = np.loadtxt("./results/energy_rbm_ngd.txt")

ngdx = ngd[:, 0]
ngdy = ngd[:, 1]


# print(len(jsx))
pyplot.plot(jsx, jsy, label = "Jastrow wave function")
# pyplot.plot(jsx, jsy, "o", markersize = 2)
pyplot.plot(rbmx, rbmy, label="RBM wave function")
pyplot.plot(ngdx, ngdy, label = "RBM wave function with Natural Gradient Descent")


pyplot.xlabel("Step")
pyplot.ylabel("Variational Energy")
pyplot.legend()

pyplot.savefig("./images/compare.png", dpi = 200)
pyplot.show()
```
